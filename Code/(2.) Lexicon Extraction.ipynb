```python
client = OpenAI(
     api_key=api_key,
     max_retries=2,
     timeout=180.0,
)
openai_query_params = {"model": "gpt-3.5-turbo-16k", "temperature": 0.25, "max_tokens": 4096}

def ask_openai(prompt: str, openai_query_params=openai_query_params):
    messages = [
        {"role": "system", "content": "You are an expert in medical information extraction."},
        {"role": "user", "content": prompt}
    ]
    response = client.chat.completions.create(
          messages=messages,
        **openai_query_params,
    )
    return response.choices[0].message.content
```

```python
base_prompt = """
###Task###
Given the above post, identify and add the physiological symptoms (excluding psychological symptoms) as key phrase separated by a comma in the column named 'Physiological Symptom Found'. Finally, Give the answers in a JSON format, where the columns are named "User name" which is the "User_name" given in the prompt, and "Physiological Symptom Found".

###Instructions###
**Do not give any other advices or writing instead the asked query.
**Do not give other result format instead json.
"""
```

```python
responses = []
```

```python

for i in range(0,10):
    post_title = data['title'][i]
    post_body = data['post'][i]
    user_name= data['authorID'][i]
    prompt="User_name:"+user_name+"\n"+"Post_title:"+post_title+"\n"+"Post_body:"+post_body +"\n"+ base_prompt
    api_response = ask_openai(prompt)
    response_dict = json.loads(api_response)
    responses.append(response_dict)
```

```python
final_responses=responses
```

```python
# Write the list of responses to a JSON file
with open('drive/My Drive/Projects/Data/PhysiologicalSymptom_GPT3response(10500).json', 'w') as json_file:
    json.dump(responses, json_file)
```

```python
import json

# Load the dataset from the provided JSON file
with open('drive/My Drive/Projects/Data/PhysiologicalSymptom_GPT3response(10500).json', 'r') as file:
     data1 = json.load(file)



# Display the first few entries to get an understanding of the dataset structure
data1_sample = data1[:5]
data1_sample
```

```python
# Checking for rows that do not have valid physiological symptoms
invalid_answers = ["None", "No", "", "N/A", "NA", "nil", "null"]

invalid_rows_count_data1 = sum(1 for entry in data1 if entry['Physiological Symptom Found'].strip() in invalid_answers)

invalid_rows_count_data1
```

```python
# Filtering out the invalid rows
filtered_data1 = [entry for entry in data1 if entry['Physiological Symptom Found'].strip() not in invalid_answers]
```

```python
# Extracting the "Physiological Symptom Found" from each entry and joining them with commas
all_symptoms_text = ", ".join([entry['Physiological Symptom Found'] for entry in filtered_data1])

# Creating the symptom related lexicon to a .txt file
file_path = "drive/My Drive/Projects/LSE-Symptomatology/Dataset/symptom_related_lexicon.txt"
with open(file_path, "w") as file:
    file.write(all_symptoms_text)


```
