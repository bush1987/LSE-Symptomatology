```python
def extract_problem_content(sample_text):
    project_id = "pcos-symptomatology"
    location_id = "us-central1"  # Replace with your selected location
    url = f"https://healthcare.googleapis.com/v1/projects/{project_id}/locations/{location_id}/services/nlp:analyzeEntities"

    data = {
        "documentContent": sample_text
    }

    response = requests.post(url, headers=headers, data=json.dumps(data))

    if response.status_code == 200:
        response_data = response.json()
        # Extract the content of "PROBLEM" type entities
        problem_contents = [mention["text"]["content"] for mention in response_data.get("entityMentions", []) if mention["type"] == "PROBLEM"]
        return problem_contents
    else:
        print("Request failed:", response.status_code, response.text)
        return []
```

```python
# Create a new column in the DataFrame to store the problem content names
df_corpus['problem_contents'] = df_corpus['sample_text'].apply(extract_problem_content)

# Display the updated DataFrame with the extracted problem contents
print(df_corpus)
```
